```python
import pandas as pd
import numpy as np
```

```python
main_df = pd.read_csv('./GLORIA_meta_and_lab.csv')
```

```python
main_df
```

```python
# All the lakes with most of the datapoints
site_names = main_df.Site_name.value_counts()[:12]
site_names
```

```python
# All the columns present in 'main_df'
main_df.columns
```

```python
# Variable containing names of all the important columns from 'main_df'
req_columns = ['Site_name','Country','Country_code','Latitude','Longitude','Date_Time_UTC','Water_body_type','Water_type','Depth','Chla','TSS','Turbidity','Secchi_depth','Chla_plus_phaeo','aCDOM440']
```

```python
# Removed all the unnecessary columns
main_df = main_df[req_columns]
main_df
```

```python
# Dropped all the rows from 'main_df' which
site_names_list = site_names.index.tolist()
main_df = main_df[main_df['Site_name'].isin(site_names_list)]
main_df
```

```python
# Converting column 'Date_Time_UTC' to 'Date' and 'Time'
main_df['Date'] = pd.to_datetime(main_df['Date_Time_UTC'], format='ISO8601', errors='coerce').dt.date
main_df['Time'] = pd.to_datetime(main_df['Date_Time_UTC'], format='ISO8601', errors='coerce').dt.time
main_df
```

```python
site_names_list
```

```python
# Generating dataframes for all the major 'Site_names' in 'main_df'
def dataframe_generator(main_df,site_names_list):
  site_dict = {}
  for site in site_names_list:
    site_dict[site] = main_df[main_df['Site_name'] == site]
  return site_dict
```

```python
site_dict = dataframe_generator(main_df,site_names_list)
```

```python
site_dict.keys()
```

```python
# Separate dataframe for each lake
Erie_df = site_dict['Lake Erie']
Gulf_of_Mexico_FL_df = site_dict['Gulf of Mexico, FL']
Geneva_df = site_dict['Lake Geneva']
IJsselmeer_df = site_dict['IJsselmeer']
Guiana_df = site_dict['Guiana']
Gulf_of_Mexico = site_dict['Gulf of Mexico']
Taihu_df = site_dict['Taihu']
English_Channel_df = site_dict['English Channel']
Garda_df = site_dict['Garda']
Kasumigaura_df = site_dict['Lake Kasumigaura']
Chesapeake_Bay_df = site_dict['Chesapeake Bay, MD']
Atlantic_Ocean_df = site_dict['Atlantic Ocean, GA']
```

```python
final_df = pd.concat([Erie_df, Gulf_of_Mexico_FL_df, Geneva_df, IJsselmeer_df,
                       Guiana_df, Gulf_of_Mexico, Taihu_df, English_Channel_df,
                       Garda_df, Kasumigaura_df, Chesapeake_Bay_df, Atlantic_Ocean_df],
                      ignore_index=True)
```
